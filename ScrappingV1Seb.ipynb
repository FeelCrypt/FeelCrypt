# 4) Cr√©ation Fonction premi√®re version NLP
## - Cr√©ation Fonction pour √©tudier les √©motions seulement sur les adverbes et adjectifs

Liste des tags possibles pour chaque mots : http://courses.washington.edu/hypertxt/csar-v02/penntable.html

```python
def EtudierEmotionTexteOnlyAdvAndAdj(url) :
    fichierTexte = open(f"{url}",encoding='utf-8' )
    text = fichierTexte.read()
    
# Suppression de la ponctuation
    text = text.replace('.','')
    text = text.replace('!','')
    text = text.replace(',','')
    text = text.replace('?','')
    text = text.replace('#','')
    text = text.replace('@','')
    
    text = text.split(' ')
    
# Suppression tous les types de mots except√© les adverbes et les adjectifs
    result_type_text = tagger.tag(text)
    liste_mots_a_conserver = []
    for i in range(len(result_type_text)) :
        if result_type_text[i][1] in ["JJ", "JJR", "JJS", "RBR", "RBS"] :
            liste_mots_a_conserver.append(result_type_text[i][0])
    
# R√©cup√©ration d'une liste d'√©motion pour chaque mots
    sentiments_par_mots = []
    for i in liste_mots_a_conserver :
        sentiments_par_mots.append(swn.senti_synsets(i))
    
# Cr√©ation d'une liste avec l'√©motion principale de chaque mots
    keepFirstOne = []
    for i in sentiments_par_mots :
        keepInMemory = i
        listI = list(i)
        if(len(listI) != 0) :
            keepFirstOne.append(listI[0])
        i = keepInMemory
    
# Somme des mots positifs et n√©gatifs
    sommePos = 0
    sommeNeg = 0
    for i in keepFirstOne :
        if(i.pos_score()>i.neg_score()) :
            sommePos = sommePos + i.pos_score()
        sommeNeg = sommeNeg + i.neg_score()
    print(f"{sommePos} {sommeNeg}")
    
# Retourne le r√©sultat
    if sommePos > sommeNeg : 
        return "Pos"
    else :
        return "Neg"
```

# 6) Version test du package VaderSentiment

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
import pandas as pd
```

```python
def EtudierEmotionTexteR(url) :
    sentiment = []
    tes = open(f"{url}",encoding='utf-8' )
    text = tes.read()
    # Suppression de la ponctuation
    text = text.replace('.','')
    text = text.replace('!','')
    text = text.replace(',','')
    text = text.replace('?','')
    text = text.replace('#','')
    
    text = text.split(' ')
    # R√©cup√©rer les √©motions mots par mots
    for tweet in text :
        score = analyser.polarity_scores(tweet)
        sentiment.append(score)
        
    tweetScore = pd.DataFrame(sentiment)
    print(tweetScore.pos.sum())
    print(tweetScore.neg.sum())
    print(tweetScore.mean())
```
