```python
import requests
import json
import csv
import pandas as pd

def get_csv_cryptocurrency_prices(currency):   
    url = f'http://api.coincap.io/v2/assets/{currency}/history?interval=d1'
    payload = {}
    headers = {}
    response = requests.request('GET', url, headers = headers, data = payload)
    currencyJson = json.loads(response.text)
    del currencyJson["timestamp"]
    currencyJsonClean = currencyJson["data"]
    return pd.DataFrame(currencyJsonClean)
```

```python
df = get_csv_cryptocurrency_prices("bitcoin")
```

### Set labels to bitcoin value dataset

```python

length = len(df)
dico = {'date': [], 
        'label': []} 

for index in range(1,length):
    value = float(df["priceUsd"][index]) - float(df["priceUsd"][index - 1])
    dico["date"].append(df["date"][index][0:10])
    dico["label"].append(1 if value > 0 else 0)
    
df_bitcoin = pd.DataFrame(dico) 
df_bitcoin.head(5)
```
