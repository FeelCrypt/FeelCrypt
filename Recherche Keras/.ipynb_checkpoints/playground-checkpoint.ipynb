```python
def get_drop_out_params(activation_fun, drop_out):
    drop_out_params = {}
    for i in range(len(activation_fun)):
        if activation_fun[i] in drop_out_params.keys():
            drop_out_params[activation_fun[i]].append(drop_out[i])
        else:
            drop_out_params[activation_fun[i]] = [drop_out[i]]
    return drop_out_params
```

```python
from itertools import permutations, combinations_with_replacement

def get_all_combinaisons(arr, item_length):
    combinaisons = list()
    for i in list(combinations_with_replacement(arr, item_length)):
        temp = permutations(i, item_length)
        combinaisons.extend(list(temp))
    return list(set(combinaisons))
```
